# G. MinMaxScaler

```python
from sklearn import preprocessing

scalers = {}
def normalize(train, test):
    # dont' scale the id
    for feature in train.columns.drop("id"):
        min_max_scaler = preprocessing.MinMaxScaler()
        train[feature] = min_max_scaler.fit_transform(train[[feature]])
        test[feature] = min_max_scaler.transform(test[[feature]])
        scalers[feature] = min_max_scaler
    return train, test
```

```python
train, test = normalize(train, test)
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
import torch.nn.functional as F
import torch.nn as nn
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
class HousesDataset(Dataset):
    def __init__(self, df):
        self.df = df
        self.x = self.df[df.columns.drop(["id", "price"])].values
        self.y = self.df[["price"]].values
        self.x = torch.tensor(self.x, dtype=torch.float32)
        self.y = torch.tensor(self.y, dtype=torch.float32)

    def __len__(self):
        return self.df.shape[0]

    def __getitem__(self, idx):
        return self.x[idx], self.y[idx]
```

```python
train_dataset = HousesDataset(train)
test_dataset = HousesDataset(test)

train_loader = DataLoader(train_dataset, batch_size=64,
                          shuffle=True, num_workers=20)
test_loader = DataLoader(test_dataset, batch_size=64,
                         shuffle=True, num_workers=20)
```

```python
for i, (data, labels) in enumerate(train_loader):
  print(data.shape, labels.shape)
  print(data,labels)
  break;
```

```python
class MLP(nn.Module):
    def __init__(self):
        super(MLP, self).__init__()
        self.fc1 = nn.Linear(20, 20)
        self.fc2 = nn.Linear(20, 15)
        self.fc3 = nn.Linear(15, 15)
        self.fc4 = nn.Linear(15, 10)
        self.fc5 = nn.Linear(10, 5)

        self.out = nn.Linear(5, 1)
    
    def forward(self, x):
        x = self.fc1(x)
        x = F.relu(x)

        x = self.fc2(x)
        x = F.relu(x)

        x = self.fc3(x)
        x = F.relu(x)

        x = self.fc4(x)
        x = F.relu(x)

        x = self.fc5(x)
        x = F.relu(x)

        x = self.out(x)
        x = F.sigmoid(x)
        return x
    def initialize_weights(self):
        # TODO what to fill
        for m in self.modules():
            if isinstance(m, nn.Linear):
                m.weight.data.fill_(1)
                if m.bias is not None:
                    m.bias.data.zero_()
```

```python
model = MLP().to(device)
model.initialize_weights()
criterion = torch.nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr = 0.01)
```

```python
data_loader = {
    "train": train_loader,
    "val": test_loader,
}
```

```python
for epoch in range(20):
    for phase in ["train", "val"]:
        if phase == "train":
            model.train(True)  # Set model to training mode
        else:
            model.train(False)  # Set model to evaluate mode
        for i, (data, labels) in enumerate(data_loader[phase], 0):
            data = data.to(device)
            labels = labels.to(device)
            optimizer.zero_grad()
            outputs = model(data)
            loss = criterion(outputs, labels)
            if phase == "train":
                loss.backward()
                optimizer.step()
```

```python
for i, (data, labels) in enumerate(test_loader):
    data = data.to(device)
    labels = labels.to(device)
    model.eval()
    pred = model(data)
    print("pred", pred)
    print("labels", labels)
    break;
```
